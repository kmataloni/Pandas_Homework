```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
file_to_load = "HeroesOfPymoli/Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)
```

```python
purchase_data
```

## Player Count

* Display the total number of players


```python
# Create new dataframe
players = len(purchase_data["SN"].value_counts())


player_count_df = pd.DataFrame({
    "Total Players": [players]
})
player_count_df
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
purchase_data.head()
```

```python
# Create item count variable
item_count = len(purchase_data["Item ID"].unique())
average_price = purchase_data["Price"].mean()
total_purchases = purchase_data["Purchase ID"].count()
total_revenue = purchase_data["Price"].sum()
```

```python
# Display the summary data
purchasingDF = pd.DataFrame({
    "Number of Unique Items": [item_count],
    "Average Price": [average_price],
    "Number of Purchases": [total_purchases],
    "Total Revenue": [total_revenue]
})


purchasingDF["Average Price"] = purchasingDF["Average Price"].round(2)

purchasingDF["Average Price"] = "$" + purchasingDF["Average Price"].astype(str)

purchasingDF["Total Revenue"] = "$" + purchasingDF["Total Revenue"].astype(str)

purchasingDF
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
# Group the data by gender
gender_demo = purchase_data.groupby("Gender")

# total players by gender
total_players_gender = gender_demo.nunique()["SN"]

# % players by gender
percentage_of_players = total_players_gender / players * 100

# gender DF
genderDF = pd.DataFrame({"Percentage of Players": percentage_of_players, 
                         "Total Count": total_players_gender})

genderDF.index.name = None

genderDF["Percentage of Players"] = genderDF["Percentage of Players"].round(2)

genderDF["Percentage of Players"] = genderDF["Percentage of Players"].astype(str) + "%"

genderDF

```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
# total purchases by gender
purchase_count = gender_demo["Purchase ID"].count()

# avg purchase by gender
avg_purchase = gender_demo["Price"].mean()

#purchase total by gender
purchase_total = gender_demo["Price"].sum()

# avg purchase total per person
avg_purchase_per_person = purchase_total / total_players_gender

# dataframe
genderDF = pd.DataFrame({"Purchase Count": purchase_count,
                         "Average Purchase Price": avg_purchase,
                         "Total Purchase Value": purchase_total,
                         "Avg Total Purchase per Person": avg_purchase_per_person})

genderDF["Average Purchase Price"] = genderDF["Average Purchase Price"].round(2)
genderDF["Total Purchase Value"] = genderDF["Total Purchase Value"].round(2)
genderDF["Avg Total Purchase per Person"] = genderDF["Avg Total Purchase per Person"].round(2)

genderDF["Average Purchase Price"] = "$" + genderDF["Average Purchase Price"].astype(str)
genderDF["Total Purchase Value"] = "$" + genderDF["Total Purchase Value"].astype(str)
genderDF["Avg Total Purchase per Person"] = "$" + genderDF["Avg Total Purchase per Person"].astype(str)

genderDF

```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
bins = [0, 9.99, 14.99, 19.99, 24.99, 29.99, 34.99, 39.99, 9999]
bin_names = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]

purchase_data["Age Group"] = pd.cut(purchase_data["Age"], bins, labels = bin_names)

age_groups = purchase_data.groupby("Age Group")

total_by_age_group = age_groups["SN"].nunique()

percent_by_age = total_by_age_group / players * 100

age_demo = pd.DataFrame({"Total Count": total_by_age_group, "Percentage of Players": percent_by_age})

age_demo["Percentage of Players"] = age_demo["Percentage of Players"].round(2)

age_demo["Percentage of Players"] = age_demo["Percentage of Players"].astype(str) + "%"

age_demo.index.name = None

age_demo
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
purchase_by_age = age_groups["Purchase ID"].count()

avg_purchase_price_by_age = age_groups["Price"].mean()

total_purchase_by_age = age_groups["Price"].sum()

age_purchase_price_percentage = total_purchase_by_age / total_by_age_group

purchaseAgeDF = pd.DataFrame({"Purchase Count": purchase_by_age,
                              "Average Purchase Price": avg_purchase_price_by_age,
                              "Total Purchase Value": total_purchase_by_age,
                              "Avg Total Purchase per Person": age_purchase_price_percentage
                             })


purchaseAgeDF["Average Purchase Price"] = purchaseAgeDF["Average Purchase Price"].round(2)
purchaseAgeDF["Total Purchase Value"] = purchaseAgeDF["Total Purchase Value"].round(2)
purchaseAgeDF["Avg Total Purchase per Person"] = purchaseAgeDF["Avg Total Purchase per Person"].round(2)

purchaseAgeDF["Average Purchase Price"] = "$" + purchaseAgeDF["Average Purchase Price"].astype(str)
purchaseAgeDF["Total Purchase Value"] = "$" + purchaseAgeDF["Total Purchase Value"].astype(str)
purchaseAgeDF["Avg Total Purchase per Person"] = "$" + purchaseAgeDF["Avg Total Purchase per Person"].astype(str)

purchaseAgeDF

```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
spender_data = purchase_data.groupby("SN")

purchase_count = spender_data["Purchase ID"].count()

avg_spender_purchase_price = spender_data["Price"].mean()

spender_purchase_total = spender_data["Price"].sum()

top_spenders = pd.DataFrame({"Purchase Count": purchase_count,
                             "Average Purchase Price": avg_spender_purchase_price,
                             "Total Purchase Value": spender_purchase_total
                            })
formattedDF = top_spenders.sort_values(["Total Purchase Value"], ascending=False).head()


formattedDF["Average Purchase Price"] = formattedDF["Average Purchase Price"].round(2)
formattedDF["Total Purchase Value"] = formattedDF["Total Purchase Value"].round(2)

formattedDF["Average Purchase Price"] = "$" + formattedDF["Average Purchase Price"].astype(str)
formattedDF["Total Purchase Value"] = "$" + formattedDF["Total Purchase Value"].astype(str)

formattedDF
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, average item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
items = purchase_data[["Item ID", "Item Name", "Price"]]

item_data = items.groupby(["Item ID", "Item Name"])

item_purchase_count = item_data["Item ID"].count()

item_purchase_value = item_data["Price"].sum()

item_price = item_purchase_value / item_purchase_count

popular_items = pd.DataFrame({"Purchase Count": item_purchase_count,
                              "Item Price": item_price,
                              "Total Purchase Value": item_purchase_value
                             })

popularFormattedDF = popular_items.sort_values(["Purchase Count"], ascending=False).head()


popularFormattedDF["Item Price"] = popularFormattedDF["Item Price"].round(2)
popularFormattedDF["Total Purchase Value"] = popularFormattedDF["Total Purchase Value"].round(2)

popularFormattedDF["Item Price"] = "$" + popularFormattedDF["Item Price"].astype(str)
popularFormattedDF["Total Purchase Value"] = "$" + popularFormattedDF["Total Purchase Value"].astype(str)

popularFormattedDF
```

## Most Profitable Items

* Sort the above table by total purchase value in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the data frame



```python
popularFormattedDF = popular_items.sort_values(["Total Purchase Value"], ascending=False).head()

popularFormattedDF["Item Price"] = popularFormattedDF["Item Price"].round(2)
popularFormattedDF["Total Purchase Value"] = popularFormattedDF["Total Purchase Value"].round(2)

popularFormattedDF["Item Price"] = "$" + popularFormattedDF["Item Price"].astype(str)
popularFormattedDF["Total Purchase Value"] = "$" + popularFormattedDF["Total Purchase Value"].astype(str)

popularFormattedDF
```
